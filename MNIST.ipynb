### comparing the model performance of components with original Columns 

```python
df = pd.read_csv('C:/Users/hp/Desktop/Python/dataset/USL-03Dec/mnist-train.csv')
df = df.head(15000)
y = df['label']
X = df.drop('label',axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state=42)
```

```python
ss = StandardScaler()

cols = X_train.columns
X_trains = ss.fit_transform(X_train)
X_trains = pd.DataFrame(X_trains, columns= cols)



X_tests = ss.transform(X_test)
X_tests =pd.DataFrame(X_tests, columns=cols)
```

```python
pca = PCA()
pca.fit(X_trains)
```

```python
pca.explained_variance_[pca.explained_variance_ >=1].shape
```

```python
pca = PCA(n_components=144)
X_trains_pc = pca.fit_transform(X_trains)
X_tests_pc = pca.transform(X_tests)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
rfc = RandomForestClassifier(n_estimators=10)

rfc.fit(X_trains_pc, y_train)

y_train_pred = rfc.predict(X_trains_pc)
y_test_pred = rfc.predict(X_tests_pc)
```

```python
## performance of random forest on train data


print(accuracy_score(y_train, y_train_pred))
print('\n')
print(confusion_matrix(y_train, y_train_pred))
```

```python
## performance of random forest on train data

print(accuracy_score(y_test, y_test_pred))
print('\n')
print(confusion_matrix(y_test, y_test_pred))
```

#### Evaluating performace on the original data

```python
rfc = RandomForestClassifier(n_estimators=10, random_state=44)
rfc.fit(X_train, y_train)
y_train_pred = rfc.predict(X_train)
y_test_pred = rfc.predict(X_test)
```

```python
print(accuracy_score(y_train, y_train_pred))
print('\n')
print(confusion_matrix(y_train, y_train_pred))
```

```python
print(accuracy_score(y_test, y_test_pred))
print('\n')
print(confusion_matrix(y_test, y_test_pred))
```

## HYPERPARAMETER TUNNIG OF RANDOM FOREST CLASSIFIER

```python
df = pd.read_csv('C:/Users/hp/Desktop/Python/dataset/USL-03Dec/mnist-train.csv')
```

```python
df  =df.head(15000)
y = df['label']
X = df.drop('label', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
cols = X.columns
ss = StandardScaler()
Xs = ss.fit_transform(X)
Xs = pd.DataFrame(Xs, columns=cols)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=148)
Xs_pc = pca.fit_transform(Xs)
```

```python
### Tunning RFC with extracted princiapl component analysis(PCA)

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

from scipy.stats import randint as sp_randint

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
```

```python
param_grid = {'n_estimators':sp_randint(10, 50),
             'max_depth':sp_randint(1, 10),
              'min_samples_split':sp_randint(2,10),
              'min_samples_leaf':sp_randint(1, 10),
              'criterion':['gini','entropy'],
              
             }


rand_search = RandomizedSearchCV(rfc, param_distributions=param_grid, cv = 3, random_state=44)

rand_search.fit(Xs_pc, y)
```

```python
rand_search.best_estimator_
```

```python
rand_search.best_params_
```

Tunning RFC with riginal data

```python
param_grid = {'n_estimators':sp_randint(10, 50),
             'max_depth':sp_randint(1, 10),
              'min_samples_split':sp_randint(2,10),
              'min_samples_leaf':sp_randint(1, 10),
              'criterion':['gini','entropy'],
              
             }


rand_search_orig = RandomizedSearchCV(rfc, param_distributions=param_grid, cv = 3, random_state=44)

rand_search_orig.fit(Xs, y)
```

```python
rand_search_orig
```

```python
rand_search_orig.best_params_
```

#### Applying the best paameters to finalize the model and evaluation

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

X_trains = ss.fit_transform(X_train)
X_tests = ss.transform(X_test)
```

```python
X_trains.shape
```

```python
X_tests.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=144)

X_trains_pc = pca.fit_transform(X_trains)
X_tests_pc = pca.transform(X_tests)
```

```python
X_trains_pc.shape
```

```python
rfc = RandomForestClassifier(**rand_search.best_params_,random_state=44) # here **rand_search.best_params_ will give the above best parameter we have got

rfc.fit(X_trains_pc, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

y_train_pred = rfc.predict(X_trains_pc)
y_test_pred = rfc.predict(X_tests_pc)
```

```python
print('accuracy of RFC on train sample')
print('accuracy', accuracy_score(y_train, y_train_pred))

print('\n','confusion_matrix')
print( confusion_matrix(y_train, y_train_pred))
```

```python
print('accuracy of RFC on test sample')
print('accuracy', accuracy_score(y_test, y_test_pred))

print('\n','confusion_matrix')
print( confusion_matrix(y_test, y_test_pred))
```
