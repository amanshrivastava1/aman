###  Clustering analysis in terms of time spent on the website

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Administrative duration vs bounce rate

X = df.iloc[:, [1, 6]]
X = pd.DataFrame(X)
```

```python
ssd = []

sscore = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    ssd.append(kmeans.inertia_)
    ss = silhouette_score(X, kmeans.labels_)
    sscore.append(ss)
```

```python
plt.plot(range(2,11), ssd)
```

```python
plt.plot(range(2,11), sscore)
```

```python
import plusmodules as pm
```

```python

pm.cluster_plot(X.values, 3)
```

#### here we can see there are 3 clusters forming. As the users spending time on a webpage is more, the chances of bouncing from the page is less. 
yellow is target users,
blue are general users,
green are uninterested user


#### we take 3 clusters as the elbow point formation is at 3 and then confirmingit with the silhoutte score, it is maximum at 3.

## Data Preprocessing

```python
# one hot encoding

df1 = pd.get_dummies(df)
```

```python
df1.columns
```

```python
df1.head()
```

```python
# label encoding of revenue


from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Revenue'] = le.fit_transform(df['Revenue'])
```

```python
df['Revenue'].value_counts()
```

```python
# taking dependent and independent variable

X = df1.drop('Revenue',axis=1)
```

```python
y = df['Revenue']
```

```python
X.columns
```

```python
# splitig the data into test and train
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
print('Training accuracy',rf.score(X_train, y_train))
print('Testing accuracy',rf.score(X_test, y_test))
```

```python
# confusion matrix

cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
#Classification report

cr = classification_report(y_test, y_pred)
print(cr)
```

### Hyperparameter Tuning

```python
param_dist = {'n_estimators':[10,20,30],'max_depth':[2,3,4,5], 'bootstrap':[True, False], 'max_features':['sqrt','log2','auto'],'criterion':['gini','entropy']}

g_rf = GridSearchCV(rf, cv = 5, param_grid=param_dist)
g_rf.fit(X_train, y_train)
```

```python
print('best parameters are',g_rf.best_params_)
```

```python
# applying the best parameters to our model

rf = RandomForestClassifier(bootstrap= True, criterion = 'gini', max_depth = 5, max_features = 'auto', n_estimators = 30)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
print('Training accuracy',rf.score(X_train, y_train))
print('Testing accuracy',rf.score(X_test, y_test))
```

```python
# confusion matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#Classification report

cr = classification_report(y_test, y_pred)
print(cr)
```
